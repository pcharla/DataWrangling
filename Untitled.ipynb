```python
p=0
my_list=[[]]

    
import requests
from bs4 import BeautifulSoup
base_url= r'https://pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s='
for page in range(0,30,10):
    print (base_url+str(page)+str('.html'))
    r= requests.get(base_url+str(page)+str('.html'))
    c = r.content
    soupy= BeautifulSoup(c,'html.parser')
    all = soupy.find_all('div',{'class':'propertyRow'})
    for item in all:
        a= item.find('h4',{'class': 'propPrice'}).text.replace('\n','')
        b=item.find_all('span',{'class':'propAddressCollapse'})[0].text
        c=item.find_all('span',{'class':'propAddressCollapse'})[1].text
        try:
            d= item.find('span',{'class':'infoBed'}).text  #find('b').text)
        except:
            d= 'None'
        try:
            e= item.find('span',{'class':'infoValueFullBath'}).text   #find('b').text)
        except:
            e= 'None'
        try:
            f= item.find('span',{'class':'infoSqFt'}).text   #find('b').text)
        except:
            f= 'None'
        try:
            g = item.find('span',{'class':'infoValueHalfBath'}).text   #find('b').text)
        except:
            g= 'None'
        for column_group in item.find_all('div',{'class':'columnGroup'}):
            for feature_group, feature_name in zip(column_group.find_all('span',{'class':'featureGroup'}),column_group.find_all('span',{'class':'featureName'})):
                if 'Lot Size' in feature_group.text:
                    try:
                        hh = feature_name.text
                    except:
                        hh = 'None'
        my_list.append ([a,b,c,d,e,f,g,hh])
```

```python
import pandas
dataframe= pandas.DataFrame(my_list,columns=['Price','Address 1','Address 2','Beds','Full Bath','Area','Half Bath','Lot Size'])
dataframe['Address']= dataframe['Address 1']+dataframe['Address 2']
```
